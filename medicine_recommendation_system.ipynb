```python
x = df.drop('prognosis',axis=1)
y = df['prognosis']
```

```python
le = LabelEncoder()
le.fit(y)
y = le.transform(y)
```

```python
y.shape
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size= 0.2,random_state=123)
```

```python
y_train
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
```

```python
rf = RandomForestClassifier(random_state=123,n_estimators=200)
```

```python
rf.fit(x_train,y_train)
```

```python
y_pred = rf.predict(x_test)
```

```python
print(r2_score(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
```

```python
train_loss = []
test_loss = []
train_accuracy = []
test_accuracy = []
n_epochs = 20
for epoch in range(1, n_epochs + 1):
    
    
    # Predict probabilities to calculate log loss
    y_train_pred_proba = rf.predict_proba(x_train)
    y_test_pred_proba = rf.predict_proba(x_test)
    
    # Calculate log loss
    train_loss.append(log_loss(y_train, y_train_pred_proba))
    test_loss.append(log_loss(y_test, y_test_pred_proba))
    
    # Calculate accuracy
    y_train_pred = rf.predict(x_train)
    y_test_pred = rf.predict(x_test)
    
    train_accuracy.append(accuracy_score(y_train, y_train_pred))
    test_accuracy.append(accuracy_score(y_test, y_test_pred))

# Plotting the training and test loss
plt.figure(figsize=(14, 6))

# Loss plot
plt.subplot(1, 2, 1)
plt.plot(range(1, n_epochs + 1), train_loss, 'bo-', label='Training Loss')
plt.plot(range(1, n_epochs + 1), test_loss, 'ro-', label='Test Loss')
plt.title('Training and Test Loss over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Log Loss')
plt.legend()
plt.grid(True)

# Accuracy plot
plt.subplot(1, 2, 2)
plt.plot(range(1, n_epochs + 1), train_accuracy, 'bo-', label='Training Accuracy')
plt.plot(range(1, n_epochs + 1), test_accuracy, 'ro-', label='Test Accuracy')
plt.title('Training and Test Accuracy over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
```

```python
def symptom_to_binary(user_symptoms):
    feature = np.zeros(len(symptoms))
    for i in user_symptoms:
        if i in symptoms:
            feature[symptoms.index(i)]=1
    
    return feature
```

```python
def predict_disease(user_symptoms):
    symp = symptom_to_binary(user_symptoms).reshape(1,-1)
    prediction = model.predict(symp)
    disease = le.inverse_transform(prediction)
    
    return disease[0]
```

```python
pre = pd.read_csv('dataset/precautions_df.csv')
medic = pd.read_csv('dataset/medications.csv')
diet = pd.read_csv('dataset/diets.csv')
workout = pd.read_csv('dataset/workout_df.csv')
discription = pd.read_csv('dataset/description.csv')
```

```python
discription
```

```python
def remidies(disease):
     
    desc = discription[discription['Disease'] == disease]['Description']
    desc = " ".join([w for w in desc])

    prec = pre[pre['Disease'] == disease][['Precaution_1', 'Precaution_2', 'Precaution_3', 'Precaution_4']]
    prec = [col for col in prec.values]

    med = medic[medic['Disease'] == disease]['Medication']
    med = [med for med in med.values]

    die = diet[diet['Disease'] == disease]['Diet']
    die = [die for die in die.values]

    wrkout = workout[workout['disease'] == disease] ['workout']
    

    return desc,med,die,wrkout,prec
```

```python
user_symptoms= ['fatigue','weight_loss','restlessness','lethargy']
Disease = predict_disease(user_symptoms)
Disease
```
